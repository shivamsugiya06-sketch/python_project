```python
movies = {
     "Avengers": {"price": 200,"total_seats":500,"seats": 5},
    "Inception": {"price": 180,"total_seats":500,"seats": 5},
        "Interstellar": {"price": 220,"total_seats":500,"seats": 5} }
bookings = []
def show_movies():
    print("\nAvailable Movies:")
    keys = list(movies.keys())
    for i in range(len(keys)):   
        name = keys[i]
        info = movies[name]
        print(f"{i+1}. {name} - Price: {info['price']} INR - Seats left: {info['seats']}")
def book_ticket():
    show_movies()
    choice = input("Enter movie number or name to book: ")
    if choice.isdigit():
        id = int(choice)
        keys = list(movies.keys())
        if 1 <= id <= len(keys):
            choice = keys[id-1]
    if choice in movies:
        if movies[choice]["seats"] > 0:
            try:
                qty = int(input("Enter number of tickets: "))
                if qty <= movies[choice]["seats"]:
                    total = qty * movies[choice]["price"]
                    movies[choice]["seats"] -= qty
                    bookings.append((choice, qty, total))  
                    print(f"\n Booked {qty} tickets for {choice}. Total = {total} INR")
                    view_bookings()  
                else:
                    print("Not enough seats available.")
            except ValueError:
                print("Please enter a valid number.")
        else:
            print("Housefull! No seats left.")
    else:
        print("Invalid movie choice.")
def view_bookings():
    print("\nYour Bookings:")
    if not bookings:
        print("No bookings yet.")
    else:
        for i in range(len(bookings)):
            b = bookings[i]
            print(f"{i+1}. Movie: {b[0]}, Tickets: {b[1]}, Amount: {b[2]} INR")
def movie_booking():
    while True:
        print("\n Movie Ticket Booking ")
        print("1. Show Movies")
        print("2. Book Ticket")
        print("3. View Bookings")
        print("4. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            show_movies()
        elif choice == "2":
            book_ticket()
        elif choice == "3":
            view_bookings()
        elif choice == "4":
            print("Thank you for using Movie Ticket Booking . Goodbye!")
            break
        else:
            print("Invalid choice! Try again.")

if __name__ == "__main__":
    movie_booking()
```
