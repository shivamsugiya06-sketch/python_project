```python
## Positive-Negative Checker using only loop

numbers = []  
results = []  
while True:  
    print("\nPositive Negative Checker")
    print("1. Enter numbers")
    print("2. Show results")
    print("3. Exit")
    choice = input("Enter your choice: ")
    if choice == "1":
        count = int(input("How many numbers you want to check? "))
        for i in range(count):
            num = int(input(f"Enter number {i+1}: "))
            numbers.append(num)
            if num > 0:
                results.append("Positive")
            elif num < 0:
                results.append("Negative")
            else:
                results.append("Zero")
    elif choice == "2":
        print("\nResults:")
        if not numbers:
            print("No numbers entered yet.")
        else:
            for i in range(len(numbers)):
                print(f"{i+1}. Number: {numbers[i]} â†’ {results[i]}")
    elif choice == "3":
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice! Try again.")
```
