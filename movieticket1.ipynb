```python
def show_movies():
    print("Available Movies:")
    print("1. Jawan")
    print("2. 3 Idiot")
    print("3. Pushpa")

def select_movie(choice):
    if choice == 1:
        return "Jawan"
    elif choice == 2:
        return "3 Idiot"
    elif choice == 3:
        return "Pushpa"
    else:
        return None

def calculate_amount(tickets, price=200):
    return tickets * price

def select_seats(tickets):
    seats = []
    print("\nPlease select your seats:")
    for i in range(tickets):
        seat_no = input(f"Seat {i+1}: ")
        seats.append(seat_no)
    return seats

print(" MOVIE TICKET BOOKING ")

show_movies()
choice = int(input("Select a movie (1-3): "))
movie = select_movie(choice)

if movie is None:
    print("Invalid choice! Please restart the program.")
    exit()

name = input("Enter your name: ")
age = int(input("Enter your age: "))   
tickets = int(input("Number of tickets: "))


if age == 18:
    print("\nNote: Since you are 18, you must watch with family.")
elif age > 18:
    print("\nYou can watch alone or with friends.")
else:
    print("\nYou are under 18, so family supervision is required.")

selected_seats = select_seats(tickets)
total_amount = calculate_amount(tickets)

print("\n BOOKING CONFIRMED ")
print(f"Name: {name}")
print(f"Age: {age}")
print(f"Movie: {movie}")
print(f"Total Tickets: {tickets}")
print(f"Seats: {', '.join(selected_seats)}")
print(f"Total Amount: ₹{total_amount}")
print("Enjoy your movie!")
```
