```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return """
    <h1>I am Home page</h1>
    <h2>I am running in Flask</h2>
    """

@app.route("/about")
def about():
    return """
    <h1>I am About page</h1>
    <h2>I am running in Flask</h2>
    """
@app.route("/contact")
def contact():
    return """
    <h1>Contact Us</h1>
    <p>Email: shivam@gmail.com</p>
    <p>Phone: 9876543210</p>
    """
if __name__ == "__main__":
    app.run(debug=True)
```
