```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt


st.title("Streamlit App of VGU")
st.text("Welcome to our dashboard")
st.header("I am header")
st.write("You can write:", 10 + 5)


name = st.text_input("Enter your name:")
age = st.number_input("Enter your age:")
st.write("Your name is:", name, "Your age is:", age)

course = st.selectbox("Enter your course:", ["BCA", "BTECH", "MCA"])

if st.button("Click me"):
    st.success("Clicked button")
    file = st.file_uploader("Upload your file")
    if file:
        content = file.read()
        st.write("File uploaded successfully !!!")


data = {"Name": ["TOM", "JACK", "hari", "om"], "marks": [10, 20, 20, 10]}
df = pd.DataFrame(data)
st.dataframe(df)


chart_data = pd.DataFrame({"marks": [10, 20, 30, 10]})
st.line_chart(chart_data)
st.bar_chart(chart_data)


subject = ["python", "c++", "java"]
marks = [20, 10, 5]
fig, ax = plt.subplots()
ax.pie(marks, labels=subject, autopct='%1.1f%%')
st.pyplot(fig)

# Bubble chart
st.title("Bubble Chart Example in Streamlit")
x = [10, 20, 30, 40, 50]
y = [5, 25, 15, 35, 10]
sizes = [100, 200, 300, 500, 400]
fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, alpha=0.5)
ax.set_title("Bubble Chart")
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")
st.pyplot(fig)


st.title("Email Input Form")
email = st.text_input("Enter your Email ID")
if st.button("Submit Email"):
    st.success(f"Your email is: {email}")

# Area chart
st.title("Area Graph Example")
x = [1, 2, 3, 4, 5]
y = [10, 30, 20, 40, 25]
fig, ax = plt.subplots()
ax.fill_between(x, y, alpha=0.5)
ax.plot(x, y)
ax.set_title("Area Chart")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
st.pyplot(fig)


st.title("Mobile Number Validation")
mobile = st.text_input("Enter Mobile Number")
if st.button("Check Mobile"):
    if mobile.isdigit() and len(mobile) == 10:
        st.success("Valid Mobile Number")
    else:
        st.error("Invalid Mobile Number (Must be 10 digits)")


st.title("Mobile Number Input (with Country Code)")
country = st.selectbox("Country Code", ["+91", "+1", "+44", "+61", "+971"])
mobile2 = st.text_input("Enter Mobile Number with Country Code")
if st.button("Submit Mobile"):
    st.success(f"Full Number: {country} {mobile2}")
```
